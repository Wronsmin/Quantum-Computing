```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
L = 32
Lattice_Size = (L, L)
periodic = True

C_G = nx.grid_graph(dim=Lattice_Size, periodic=periodic)
```

```python
C_G.number_of_edges()
```

```python
for node in C_G.nodes:
    x, y = node
    
    # diagonals on right
    x1, y1 = (x+1) % L, (y+1) % L
    C_G.add_edge((x, y), (x1, y1))
    
    x1, y1 = (x-1) % L, (y+1) % L
    C_G.add_edge((x, y), (x1, y1))
    
    # diagonals on left
    x1, y1 = (x+1) % L, (y-1) % L
    C_G.add_edge((x, y), (x1, y1))
    
    x1, y1 = (x-1) % L, (y-1) % L
    C_G.add_edge((x, y), (x1, y1))
    
```

```python
nx.draw(G=C_G, node_size=10)
```
