```python
import dwave_networkx as dnx
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
width = 5.949973986 # in inches
ratio = (np.sqrt(5.0) - 1.0) / 2.0

cm = 1/2.54 # cm to inches
width = 12 * cm

grid = plt.GridSpec(1, 2)

fig = plt.figure(figsize=(width, 0.5*width * ratio))
fig.set_tight_layout({'pad': 0})
```

```python
G = dnx.chimera_graph(1)

dnx.draw_chimera(G, node_size=450, with_labels=True, width=2, edgecolors='k')
```

```python
G = dnx.chimera_graph(1)

Chim_ax = fig.add_subplot(grid[0, 0])
Tri_ax = fig.add_subplot(grid[0, 1])

dnx.draw_chimera(G, node_size=450, with_labels=True, width=2, edgecolors='k', node_color='lightgray', ax=Chim_ax)
```

```python
G = nx.Graph()

pos = {'A': [0, 0],
       'B': [1, 1],
       'C': [2, 0]}

G.add_edges_from([('A','B'), ('B','C'), ('C','A')])

nx.draw(G, pos=pos, with_labels=True, node_size=450, width=2, edgecolors='k', node_color='lightgray', ax=Tri_ax)
```

```python
fig = plt.figure(figsize=(width, 0.5 * width * ratio))
fig.set_tight_layout({'pad': 0})

pos = {5: [0, 1],
       2: [1, 1], 
       6: [1, 0], 
       1: [0, 0]}

S1_ax = fig.add_subplot(grid[0, 0])
S2_ax = fig.add_subplot(grid[0, 1])
```

```python
G1 = nx.Graph()
G1.add_edges_from([(1, 5), (5, 2), (2, 6), (6, 1)])
```

```python
nx.draw(G1, pos=pos, with_labels=True, node_size=250, width=2, edgecolors='k', node_color='lightgray', ax=S1_ax)
nx.draw(G1, pos=pos, with_labels=True, node_size=250, width=2, edgecolors='k', node_color='lightgray', ax=S2_ax)
```
