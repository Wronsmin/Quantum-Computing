```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
width = 5.949973986 # in inches
ratio = (np.sqrt(5.0) - 1.0) / 2.0
```

```python
def f(x, c):
    return 0.5*x**4-x**2-c*x
```

```python
grid = plt.GridSpec(1, 10)

fig = plt.figure(figsize=(width, width * ratio))
fig.set_tight_layout({'pad': 0})
v_row_ax = fig.add_subplot(grid[0, 0])
superstate_ax = fig.add_subplot(grid[0, 1:4])
double_well_ax = fig.add_subplot(grid[0, 4:7])
dw_mag_ax = fig.add_subplot(grid[0, 7:])


v_row_ax.annotate('High Energy', (0, 0.01),
            xytext=(0.0, 0.95),
            arrowprops=dict(facecolor='black',
                            shrink=0.1, 
                            width=1,
                            headwidth=4, 
                            headlength=3),
            fontsize=11,
            horizontalalignment='center', verticalalignment='top')
v_row_ax.annotate('Low Energy', (0, 0.04), 
                  fontsize=11, horizontalalignment='center')
v_row_ax.set_axis_off()


x = np.linspace(-3, 3, 100)
superstate_ax.plot(x, 0.5*x**2, 'k')
superstate_ax.annotate('Superposition\nstate', (0, 0.1),
            xytext=(0.0, 1),
            arrowprops=dict(facecolor='black', 
                            shrink=0.1, 
                            width=1,
                            headwidth=4, 
                            headlength=3),
            fontsize=5,
            horizontalalignment='center', verticalalignment='top')
superstate_ax.annotate('(a)', (0, 0.5*3**2), 
                  fontsize=11, horizontalalignment='center')

superstate_ax.set_axis_off()


x = np.linspace(-1.5, 1.5, 100)

double_well_ax.plot(x, f(x, 0.0), 'k')
double_well_ax.annotate('(b)', (0, f(-1.5, 0.0)), 
                  fontsize=11, horizontalalignment='center')
double_well_ax.annotate('0', (-1, f(0, 0.0)*0.05), 
                  fontsize=11, horizontalalignment='center',
                  weight='bold')
double_well_ax.annotate('1', (1, f(0, 0.0)*0.05), 
                  fontsize=11, horizontalalignment='center', 
                  weight='bold')
double_well_ax.set_axis_off()


x = np.linspace(-1.55, 1.676, 100)

dw_mag_ax.plot(x, f(x, 0.2), 'k')
dw_mag_ax.annotate('(c)', (0, f(-1.5, 0.2)), 
                  fontsize=11, horizontalalignment='center')

dw_mag_ax.annotate('0', (-1, f(0, 0.2)*0.05), 
                  fontsize=11, horizontalalignment='center',
                  weight='bold')
dw_mag_ax.annotate('1', (1, f(0, 0.2)*0.05), 
                  fontsize=11, horizontalalignment='center', 
                  weight='bold')
dw_mag_ax.annotate('Higher\nprobability\nof lower state', (1, f(0.945649, 0.2)), 
                   xytext=(-1,f(0.945649, 0.2) + 0.1),
                   arrowprops=dict(facecolor='black', 
                            shrink=0.1, 
                            width=0.5,
                            headwidth=4, 
                            headlength=3),
                  fontsize=5, horizontalalignment='center', verticalalignment='bottom')
dw_mag_ax.set_axis_off()

#fig.savefig("prova.svg")
```

```python
fig = plt.figure(figsize=(width, 0.7*width * ratio))
fig.set_tight_layout({'pad': 0})

x = np.linspace(0, 1)
plt.plot(x, 1-x, label=r'$A(t/t_{max}$)')
plt.plot(x, x, 'r', label=r'$B(t/t_{max}$)')
plt.xlabel(r'$t/t_{max}$')
plt.ylabel(r'$Energy$ $[arbitrary$ $units]$')
plt.legend()
plt.grid()
plt.savefig("linear_adiabatic_annealing.pdf")
```

```python
import pandas as pd 

AS = pd.read_excel("Advantage_system6_1.xlsx")

GHztoJ = 6.62606957030463E-25

t = AS['s'].values
A = AS['A(s) (GHz)'].values * GHztoJ
B = AS['B(s) (GHz)'].values * GHztoJ
```

```python
fig = plt.figure(figsize=(width,0.7* width * ratio))
fig.set_tight_layout({'pad': 0})

plt.plot(t, A, label=r'$A(t/t_{max}$)')
plt.plot(t, B, 'r', label=r'$B(t/t_{max}$)')
plt.xlabel(r'$t/t_{max}$')
plt.ylabel(r'$Energy$ $[Joules]$')
plt.legend()
plt.grid()
plt.savefig("non_linear_adiabatic_annealing.pdf")
```
