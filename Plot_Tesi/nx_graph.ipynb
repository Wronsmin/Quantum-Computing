```python
import numpy as np 
import networkx as nx
import matplotlib.pyplot as plt


width = 5.949973986 # in inches
ratio = (np.sqrt(5.0) - 1.0) / 2.0

fig = plt.figure(figsize=(0.6*width, 0.6*0.5*width * ratio))
fig.set_tight_layout({'pad': 0})

G = nx.Graph()

G.add_edge(r"$q_1$", r"$q_2$")

nx.draw(G, with_labels=True, node_size=300, width=2, edgecolors='k', node_color='lightgray')
#plt.savefig("BQM.svg")
```

```python
pos = {1: [0, 5],
       2: [5, 5], 
       3: [5, 0], 
       4: [0, 0]}
```

```python
G1 = nx.DiGraph()
G1.add_edges_from([(1, 2), (2, 1),(2, 3), (3, 4), (4, 3),(1, 1), (3, 3)])
```

```python
nx.draw(G1, pos=pos, with_labels=True, node_size=2000, width=2, edgecolors='k', node_color='lightgray', connectionstyle='arc3, rad = 0.1')
```

```python
G2 = nx.DiGraph()
G2.add_edges_from([(1, 2), (2, 1),(2, 3), (3, 4), (4, 3),(1, 1), (3, 3), (4, 1)])

nx.draw(G2, pos=pos, with_labels=True, node_size=2000, width=2, edgecolors='k', node_color='lightgray', connectionstyle='arc3, rad = 0.1')
```
